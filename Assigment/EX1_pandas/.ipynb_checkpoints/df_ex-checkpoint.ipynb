```python
import pandas as pd
```

```python
alist = ['Mouse', 'Giraffe', 'Tiger', 'Dog', 'Hippopotamus', 'Moose', 'Bear']
animals = pd.Series(alist)
animals
```

```python
names = ['Chris', 'Kevin', 'Bob', 'Vinod', 'Daniel', 'Fil', 'Stephanie']
owners = pd.Series(names)
owners
```

```python
frame = {'animals': animals, 'owners': owners}

df = pd.DataFrame(frame)

df
```

```python
df.columns
```

```python
df.index
```

```python
type(df['animals'])
```

```python
df.iloc[0]
```

```python
midx = pd.MultiIndex(levels=[['lama', 'cow', 'falcon'],
                             ['speed', 'weight']],
                     codes=[[0, 0, 1, 1, 2, 2],
                            [0, 1, 0, 1, 0, 1]])

midx
```

```python
df = pd.DataFrame(index=midx, columns=['big', 'small'],
                   data=[[45, 30], [200, 100], [30, 20], [250, 150], [320, 250],[1, 0.8]])

df
```

```python
df.drop(index='cow', columns='small')
```

```python
df.drop(index='weight', level=1)
```

## Part 1 Task 1

Given the data below, implement each tasks. 

<img src="img/df.png" style="height:200px"> 

(1) Create a DataFrame.

(2) Access the first row data and print it.

(3) Access the first column data (indexed by 'dept') and print it.

(4) Create a new DataFrame by dropping the 2nd column data (indexed by 'floor') from the current DataFrame.

```python
import pandas as pd

# (1) Create the DataFrame
data = {
    "dept": ["CS", "Bio", "Chem", "Phy", "Eng"],
    "floor": [5, 3, 3, 2, 4]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# (2) Access and print the first row
print("\nFirst row:")
print(df.iloc[0])
```

```python
# (3) Access and print the first column ('dept')
print("\nFirst column (dept):")
print(df["dept"])
```

```python
# (4) Create a new DataFrame by dropping the 'floor' column
df_new = df.drop(columns=["floor"])
print("\nDataFrame after dropping 'floor' column:")
print(df_new)
```
