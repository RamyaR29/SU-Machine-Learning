```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
```

```python
lifesat.head()
```

```python
# Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Ensure lifesat DataFrame is loaded (assuming it's already in the kernel)
# If not, the following lines would be needed:
# data_root = "https://github.com/ageron/data/raw/main/"
# lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")

# Create the scatter plot
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction", figsize=(12, 8))
plt.axis([23_500, 62_500, 4, 9])

# Add country names as annotations
for i, country in enumerate(lifesat['Country']):
    plt.text(lifesat['GDP per capita (USD)'][i] + 500, # Adjust x position slightly
             lifesat['Life satisfaction'][i] - 0.1,  # Adjust y position slightly
             country, fontsize=8, alpha=0.7)

plt.title('Life Satisfaction vs. GDP per Capita with Country Names')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Life satisfaction')
plt.show()
```

```python
# Select a linear model
model = LinearRegression()
```

```python
# Train the model
model.fit(X, y)
```

```python
# Make a prediction for Cyprus
X_new = [[33_442.8]]  # Cyprus' GDP per capita in 2020
print(model.predict(X_new)) # output: [[6.30165767]] # model based training
```

```python
from sklearn.neighbors import KNeighborsRegressor
kmeansmodel = KNeighborsRegressor(n_neighbors=3)
```

```python
kmeansmodel.fit(X, y)
# Make a prediction for Cyprus
X_new = [[33_442.8]]  # Cyprus' GDP per capita in 2020
print(kmeansmodel.predict(X_new)) # output: [[6.30165767]] # instance based training
```
