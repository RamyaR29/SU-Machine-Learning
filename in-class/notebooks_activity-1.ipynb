Ex 3: Prompt the user to enter 10 integers and store them in a list, then print:

Mean (average)
Min and Max
Count of numbers > 50
All odd numbers
All numbers divisible by 3
The list sorted (ascending and descending)
The list reversed

```python
import math
list = []
great = 0
for i in range(10):
    num = int(input("Enter number:"))
    if num > 50:
        great = great + 1
    list.append(num)
mean = sum(list)/len(list)
print("mean: ", mean)
print("min :", min(list))
print("max :", max(list))
print("Greater than 50 :", great)
print ("Odd numbers: ",[num for num in list if num % 2 != 0])
print ("Numbers % 3: ",[num for num in list if num % 3 == 0])
print("Ascending : ", sorted(list))
print("Descending : ", sorted(list, reverse = True))

reversed_list = list[::-1]
print("Reversed list :", reversed_list)
```
